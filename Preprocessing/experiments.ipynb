```python
import sys
sys.path.append('/Users/1/Dropbox/')
sys.path.append('/Users/1/Dropbox/Features')

from scipy.io import wavfile as wav
import re
import os
import glob
import numpy as np
from file_to_mixture import unite, get_parts
from result_function import likelihood
import MFCC
from MFCC import delta, get_mfcc
from sklearn import mixture
import matplotlib.pyplot as plt
import scipy
%matplotlib inline


def midpoint(scores, N):
    '''Вычисляет серединную точку между границами спикера и imposter model,
       N - количество спикеров в imposter model'''
    midpoint = (min(scores[:-N]) + max(scores[-N:])) / 2
    return midpoint 


def errors(scores, threshold, N, single=None):
    '''Процент ошибок первого и второго рода,
       N - количество спикеров в imposter mode'''
    if not single:
        i = [ ((len([i for i in scores[j][:-N] if i <= threshold])), len(scores[j][:-N])) for j in range(len(scores)) ]
        ii = [ ((len([i for i in scores[j][-N:] if i >= threshold])), len(scores[j][-N:])) for j in range(len(scores)) ]  
        
        typeI, typeII = [sum( [x[i][0] for i in range(len(x))] ) / sum( [x[j][1] for j in range(len(x))] ) for x in (i, ii)]

    if single:
        typeI = (len([i for i in scores[:-N] if i <= threshold])) / len(scores[:-N])
        typeII = (len([i for i in scores[-N:] if i >= threshold])) / len(scores[-N:]) 
        
    return typeI * 100, typeII * 100
```
